```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pytesseract
```

```python
pytesseract.pytesseract.tesseract_cmd = r'D:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
img = cv2.imread('imgs/texto.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
print(pytesseract.image_to_string(img))
```

```python
_, axis = plt.subplots(1, 2)
img = cv2.imread('imgs/bh.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
axis[0].imshow(img)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img = cv2.threshold(img, 200, 200, cv2.THRESH_BINARY)[1]
axis[1].imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2BGR))
plt.show()
print(pytesseract.image_to_string(img))
```

```python
_, axis = plt.subplots(1, 2)
img = cv2.imread('imgs/senhor dos aneis.png', cv2.IMREAD_GRAYSCALE)
axis[0].imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2BGR))
img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
img = cv2.GaussianBlur(img, (3, 3), 5)
# img = cv2.medianBlur(img, 1)
img = cv2.threshold(img, 100, 255, cv2.THRESH_TOZERO)[1]
axis[1].imshow(img)
print(pytesseract.image_to_string(img))
```
